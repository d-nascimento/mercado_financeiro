```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
import warnings
warnings.filterwarnings('ignore')
```

```
ibov = yf.Ticker('^BVSP')
```

```
ibov.info['currency']
```

```
ibov_df = ibov.history()
```

```
ibov_df.head()
```

```
plt.figure(figsize=[13,8])
plt.plot(ibov_df['Close']);
```

```
tsla = yf.Ticker("TSLA").history(period='max',actions=True)
tsla.head()
```

```
plt.figure(figsize=[13,8])
plt.plot(tsla.Close)
plt.xlabel('Year')
plt.ylabel('Price $');
```

```

plt.figure(figsize=[13,8])
plt.plot(tsla.Close[tsla.Close.index > '2020'],label='Close',color='blue')
plt.plot(tsla.Close[tsla.Close.index > '2020'].rolling(21).mean(),label='MM21',color='black')
plt.plot(tsla.Close[tsla.Close.index > '2020'].rolling(42).mean(),label='MM42',color='red')
plt.xlabel('Year')
plt.ylabel('Price $')
plt.title('TSLA | Histórico de Preços')
plt.legend();
```
